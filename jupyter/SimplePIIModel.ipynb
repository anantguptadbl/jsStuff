```python
# We will create a model that will identify PAN Number in the data

input_text = ['Anant Gupta with the SSN Number of 518-001-9987 wants to apply for a new card',
             'We would also like to registe ra complaint for the safety of data',
             'This is the best service what we have got with the new card',
             'Anant Gupta with the SSN Number wants to apply for a new card',
             'Sweta Agarwal also wants to apply for a new card']

from gensim.models import Word2Vec
from tensorflow import keras

model = Word2Vec(sentences=input_text, window=2, min_count=1, workers=4)
model.wv.get_keras_embedding(train_embeddings=False)
```

```python
# We will write a very simple model.. something random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflowjs as tfjs
import os

from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers, losses
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Model
from tensorflow import keras

input_img = keras.Input(shape=(2,))
# The first attribute is presence of word SSN
# The second is the presence of the SSN text
encoded = layers.Dense(32, activation='relu')(input_img)
encoded = layers.Dense(64, activation='relu')(encoded)
decoded = layers.Dense(128, activation='relu')(encoded)
decoded = layers.Dense(64, activation='relu')(decoded)
decoded = layers.Dense(1, activation='sigmoid')(decoded)
autoencoder = keras.Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='mse')

train_data = np.array([
    [1,1],
    [0,1],
    [0,0],
    [1,0],
    [0,0]
])

autoencoder.fit(train_data, np.array([1,0,0,0,0]),
                epochs=100,
                shuffle=True,
                verbose=1
                )

# Keras way to save
tf.keras.models.save_model(
    autoencoder, "pii_model.h5", overwrite=True, include_optimizer=True, save_format='h5'
)

!tensorflowjs_converter --input_format=keras pii_model.h5 tfjs_pii_model

autoencoder.predict([[1,0]])
```
